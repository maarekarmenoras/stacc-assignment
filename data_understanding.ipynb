# Data understanding

### Import packages and dataset

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, geom_line
```

```python
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
iris = pd.read_csv(url)
```

### NaN-values, duplicates and outliers

#### NaN values per column

```python
iris.isna().sum(axis=0)
```

#### Number of duplicates

```python
iris.duplicated().sum()
```

#### Visual value distributions

```python
(
    ggplot(data=iris)
    + geom_bar(aes(x='sepal_length', color='species'))
)
```

```python
(
    ggplot(data=iris)
    + geom_bar(aes(x='sepal_width', color='species'))
)
```

```python
(
    ggplot(data=iris)
    + geom_bar(aes(x='petal_length', color='species'))
)
```

```python
(
    ggplot(data=iris)
    + geom_bar(aes(x='petal_width', color='species'))
)
```

### Conclusion

The iris dataset is already fairly clean with no NaN values and minimal duplicates. There are exactly 50 samples of each species, so no over- or undersampling is needed. Based on the visualisations, there are some outliers, but it is important to take into account the value distributions for each species and not necessarily the whole, especially in the case of petal width and length. Sepal width and length for any species are harder to incur from this type of plot, but is kept around for looking at overall trends and consistency.
